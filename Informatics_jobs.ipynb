```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
from collections import Counter
import torch
import nltk
from nltk.tokenize import sent_tokenize
import gc
import re

# Download NLTK data for sentence tokenization
nltk.download('punkt')

# Load your DataFrame (replace with your actual data loading method)
# Example:
# filtered_df = pd.read_csv('informatics_jobs.csv')

# Specify the model name
model_name = "jjzha/jobberta-base"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Initialize the NER pipeline without truncation parameter
ner_pipeline = pipeline(
    "ner",
    model=model,
    tokenizer=tokenizer,
    grouped_entities=True,
    device=-1  # Use CPU
)

def extract_skills_certifications_combined(description):
    """
    Extracts skills and certifications from a job description using both NER and regex methods.
    
    Parameters:
    - description (str): The job description text.
    
    Returns:
    - skills (list): A list of extracted skills.
    - certifications (list): A list of extracted certifications.
    """
    skills = []
    certifications = []
    
    # Split the description into sentences
    sentences = sent_tokenize(description)
    
    for sentence in sentences:
        # Encode the sentence with truncation
        encoded = tokenizer.encode_plus(
            sentence,
            truncation=True,
            max_length=512,
            return_tensors='pt'
        )
        
        # Decode back to truncated sentence
        truncated_sentence = tokenizer.decode(encoded['input_ids'][0], skip_special_tokens=True)
        
        # Perform NER on the truncated sentence
        ner_results = ner_pipeline(truncated_sentence)
        
        for entity in ner_results:
            label = entity['entity_group'].lower()
            word = entity['word'].strip()
            
            if 'skill' in label:
                skills.append(word)
            elif 'certification' in label or 'license' in label:
                certifications.append(word)
            elif 'org' in label or 'product' in label:
                # Optionally, categorize ORG or PRODUCT entities as certifications or specific skills
                # Modify this based on inspection results
                certifications.append(word)
        
        # Additionally, use regex to find skills and certifications
        # Define regex patterns
        cert_patterns = [
            r'certifications?:\s*([\w\s,/-]+)', r'certificates?:\s*([\w\s,/-]+)',
            r'required certifications?:\s*([\w\s,/-]+)', r'License?:\s*([\w\s,/-]+)', r'licensure?:\s*([\w\s,/-]+)',
            r'required license?:\s*([\w\s,/-]+)'
            r'relevant certifications?:\s*([\w\s,/-]+)',  r'qualifications?:\s*([\w\s,/-]+)'
        ]
        skill_patterns = [
            r'skills?:\s*([\w\s,/-]+)', r'abilities?:\s*([\w\s,/-]+)',
            r'required skills?:\s*([\w\s,/-]+)',
            r'knowledge?:\s*([\w\s,/-]+)', r'required knowledge?:\s*([\w\s,/-]+)',
            r'relevant skills?:\s*([\w\s,/-]+)'
        ]
        
        # Extract certifications using regex
        for pattern in cert_patterns:
            matches = re.findall(pattern, sentence, re.IGNORECASE)
            for match in matches:
                certs = re.split(r',|/|&|\n', match)
                certifications.extend([cert.strip() for cert in certs if cert.strip()])
        
        # Extract skills using regex
        for pattern in skill_patterns:
            matches = re.findall(pattern, sentence, re.IGNORECASE)
            for match in matches:
                skl = re.split(r',|/|&|\n', match)
                skills.extend([skill.strip() for skill in skl if skill.strip()])
    
    return skills, certifications

def clean_items(items):
    """
    Cleans and standardizes extracted items.
    
    Parameters:
    - items (list): List of extracted items.
    
    Returns:
    - cleaned_items (list): Cleaned and standardized list.
    """
    cleaned = []
    for item in items:
        # Remove any trailing punctuation
        item = item.rstrip('.,;:')
        # Standardize casing: keep acronyms uppercase, title case others
        if item.isupper():
            cleaned.append(item)
        else:
            cleaned.append(item.title())
    return cleaned

# Initialize Lists to Store Extracted Data
all_skills = []
all_certifications = []

# Define Batch Size
batch_size = 10
num_rows = len(filtered_df)
num_batches = (num_rows // batch_size) + 1

# Iterate Through Each Batch
for batch_num in range(num_batches):
    start_idx = batch_num * batch_size
    end_idx = start_idx + batch_size
    batch_df = filtered_df.iloc[start_idx:end_idx]
    
    for index, row in batch_df.iterrows():
        description = row['description']
        
        # Ensure the description is a string
        if not isinstance(description, str):
            continue  # Skip if description is not a string
        
        try:
            # Extract skills and certifications using combined method
            skills, certifications = extract_skills_certifications_combined(description)
            
            # Append the extracted data to the respective lists
            all_skills.extend(skills)
            all_certifications.extend(certifications)
        
        except Exception as e:
            print(f"Error processing row {index}: {e}")
            continue  # Skip to the next row in case of an error
    
    print(f"Processed batch {batch_num + 1} / {num_batches}")
    
    # Clear variables to free memory
    del batch_df
    gc.collect()

# Clean the extracted skills and certifications
clean_skills = clean_items(all_skills)
clean_certifications = clean_items(all_certifications)

# Remove duplicates by converting to sets
unique_skills = sorted(set([skill for skill in clean_skills if skill]))
unique_certifications = sorted(set([cert for cert in clean_certifications if cert]))

# Display Aggregated Results
print("\nCertifications required across jobs:")
for cert in unique_certifications:
    print(f"- {cert}")

print("\nSkills required across jobs:")
for skill in unique_skills:
    print(f"- {skill}")

# (Optional) Frequency Analysis
skill_counter = Counter(clean_skills)
certification_counter = Counter(clean_certifications)

# Convert counters to DataFrames for better visualization
skill_freq_df = pd.DataFrame(skill_counter.items(), columns=['Skill', 'Frequency']).sort_values(by='Frequency', ascending=False)
cert_freq_df = pd.DataFrame(certification_counter.items(), columns=['Certification', 'Frequency']).sort_values(by='Frequency', ascending=False)

# Display the top 20 skills
print("\nTop 20 Skills:")
print(skill_freq_df.head(20))

# Display the top 20 certifications
print("\nTop 20 Certifications:")
print(cert_freq_df.head(20))

# Save the results to CSV files
pd.DataFrame(unique_skills, columns=['Skill']).to_csv('unique_skills.csv', index=False)
pd.DataFrame(unique_certifications, columns=['Certification']).to_csv('unique_certifications.csv', index=False)
skill_freq_df.to_csv('skill_frequencies.csv', index=False)
cert_freq_df.to_csv('certification_frequencies.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Assuming skill_df is already defined with 'Skill & Knowledge' and 'Label' columns

# 1. Count the occurrences of each label
label_counts = skill_df['Label'].value_counts()

# 2. Display the counts
print("Label Counts:\n", label_counts)

# 3. Plot a bar chart
plt.figure(figsize=(12,8))
ax = sns.barplot(x=label_counts.index, y=label_counts.values, palette='magma')

plt.title('Distribution of Labels in skill_df', fontsize=18)
plt.xlabel('Label', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Add count annotations
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=12, color='black', 
                xytext=(0,5),
                textcoords='offset points')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Handling Top N Labels (e.g., Top 10)
top_n = 5
top_labels = label_counts.nlargest(top_n)

plt.figure(figsize=(12,8))
ax = sns.barplot(x=top_labels.index, y=top_labels.values, palette='viridis')

plt.title(f'Top {top_n} Most Common Labels in skill_df', fontsize=18)
plt.xlabel('Label', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Add count annotations
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=12, color='black', 
                xytext=(0,5),
                textcoords='offset points')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Plot a pie chart for label proportions
plt.figure(figsize=(8,8))
colors = sns.color_palette('pastel')[0:len(label_counts)]

label_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=colors, figsize=(8,8))
plt.title('Proportion of Labels in skill_df', fontsize=16)
plt.ylabel('')  # Hide y-label
plt.tight_layout()
plt.show()
```
